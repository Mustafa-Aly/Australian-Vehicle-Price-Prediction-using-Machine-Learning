#**Bivarient Analysis**

Corrolation between Numeric Columns

what the 15 most expensive cars ?

```python
df.groupby(['Brand & Model'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False)
```

```python
px.histogram(x=df.groupby(['Brand & Model'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False).head(15)['Brand & Model'],
             y=df.groupby(['Brand & Model'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False).head(15)['Price'])
```

what the 15 most cheapest cars ?

```python
px.histogram(x=df.groupby(['Brand & Model'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=True).head(15)['Brand & Model'],
             y=df.groupby(['Brand & Model'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=True).head(15)['Price']).update_xaxes(categoryorder='total descending')
```

What is more expensive automatic cars or manuel ?

```python
df.groupby(['Transmission'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False)
```

```python
px.histogram(x=df.groupby(['Transmission'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False).head(15)['Transmission'],
             y=df.groupby(['Transmission'])['Price'].mean().reset_index().sort_values(by=['Price'],ascending=False).head(15)['Price'])
```

Is Price depending on Colour of the Car?

```python
df[df['Year']==2020].groupby(['Brand & Model','ColourExtInt'])['Price'].mean().reset_index()
```

```python
px.histogram(x=df[df['Year']==2020].groupby(['Brand & Model','ColourExtInt'])['Price'].mean().reset_index().head(50)['Brand & Model'],
             color=df[df['Year']==2020].groupby(['Brand & Model','ColourExtInt'])['Price'].mean().reset_index().head(50)['ColourExtInt'],
             y=df[df['Year']==2020].groupby(['Brand & Model','ColourExtInt'])['Price'].mean().reset_index().head(50)['Price'],
             barmode='group')
```

What Colour is the most selling in Last 3 years?

```python
df[df['Year']>2020].groupby(['ColourExtInt'])['ColourExtInt'].count()
```

```python
px.histogram(x=df[df['Year']>2020].groupby(['ColourExtInt'])['ColourExtInt'].count().index,y=df[df['Year']>2020].groupby(['ColourExtInt'])['ColourExtInt'].count().values).update_xaxes(categoryorder='total descending')
```

Does Fuel consumption [ / 100km] affect on the Price ?

```python
df[df['FuelConsumption']==20.7]
```
