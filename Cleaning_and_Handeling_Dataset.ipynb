Extract the first color from the column and remove false inputs

```python
def extract_color(x):
  x=x.split('/')
  if x[0].strip()=='-':
    return np.nan
  elif  x[0].split(' ')[0] in (['1','2','3','4','5','6','7','8','9','0']) :
    return np.nan
  else:
    return x[0].strip()
```

```python
extract_color('- / -')
```

```python
extract_color('5 years / 200,000 km')
```

```python
extract_color('Silver / Leather Charcoal/Charcoal/Blond Headlining')
```

```python
df['ColourExtInt']=df['ColourExtInt'].apply(extract_color)
```

```python
df['ColourExtInt'].value_counts()
```
