```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline , make_pipeline

from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
from sklearn.preprocessing import OneHotEncoder
import category_encoders as ce

from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

from sklearn.decomposition import PCA
from sklearn.feature_selection import SequentialFeatureSelector,SelectFromModel

from sklearn.model_selection import cross_validate,cross_val_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV

```

```python
for col in df.select_dtypes('object').columns:
  print(col,' ',df[col].nunique())
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
x
```

```python
y
```

```python
px.histogram(np.log(y))
```

```python
log_y=np.log(y)
```

```python
px.histogram(log_y)
```

```python
log_col=['Kilometres']
ohe_col=['UsedOrNew','Transmission','DriveType']
be_col=['FuelType','ColourExtInt','BodyType','Brand & Model','State']
```

```python
# log_fun= FunctionTransformer(np.log1p)
```

```python
transform=ColumnTransformer(transformers=[
    ('log',FunctionTransformer(np.log1p),log_col),
    ('Ohe',OneHotEncoder(sparse_output=False,drop='first'),ohe_col),
    ('Be',ce.BinaryEncoder(),be_col)
],remainder='passthrough')
```

```python
transform
```

```python
models=[
    ('lr',LinearRegression()),
    ('Lasso',Lasso()),
    ('Ridge',Ridge()),
    ('SVC',SVR()),
    ('Knn',KNeighborsRegressor()),
    ('DT',DecisionTreeRegressor()),
    ('Bagging',BaggingRegressor()),
    ('RT',RandomForestRegressor()),
    ('XGB',XGBRegressor())
]
```

**compare the result from Y and Log(Y)**

```python
for model in models:
  print(model[0])
  PL=make_pipeline(transform,
                   StandardScaler(),

                   model[1])

  cv=cross_validate(estimator=PL,X=x,y=y,cv=7,return_train_score=True,scoring=['r2','neg_mean_squared_error'])
  print(model[0],'train r2 score = ',cv['train_r2'].mean())
  print(model[0],'test r2 score = ',cv['test_r2'].mean())
  print(model[0],'train mean_squared_error = ',cv['train_neg_mean_squared_error'].mean())
  print(model[0],'test mean_squared_error = ',cv['test_neg_mean_squared_error'].mean())
  print('-'*30)
```

```python
for model in models:
  print(model[0])
  PL=make_pipeline(transform,
                   StandardScaler(),

                   model[1])

  cv=cross_validate(estimator=PL,X=x,y=log_y,cv=7,return_train_score=True,scoring=['r2','neg_mean_squared_error'])
  print(model[0],'train r2 score = ',cv['train_r2'].mean())
  print(model[0],'test r2 score = ',cv['test_r2'].mean())
  print(model[0],'train mean_squared_error = ',cv['train_neg_mean_squared_error'].mean())
  print(model[0],'test mean_squared_error = ',cv['test_neg_mean_squared_error'].mean())
  print('-'*30)
```

XGBoost parameter Hypertuning


```python
PL=make_pipeline(transform,
                  StandardScaler(),

                  XGBRegressor())

XGB_params=[
    {
        'xgbregressor__learning_rate':[0.05,0.1,0.15,0.20,0.25,0.30],
        'xgbregressor__max_depth':[3,4,5,6,8,10,12,15],
        'xgbregressor__min_child_weight':[1,3,5,7],
        'xgbregressor__gamma':[0.0,0.1,0.2,0.3,0.4],
        'xgbregressor__colsample_bytree':[0.3,0.4,0.5,0.7]

    }
]

XGBRegressor_RS = RandomizedSearchCV(PL, XGB_params,n_iter=80, scoring='r2', cv=5,n_jobs=-1)
XGBRegressor_RS.fit(x,log_y)
```

```python
print("Best value for lambda : ",XGBRegressor_RS.best_params_)
print("Best score for cost function: ", XGBRegressor_RS.best_score_)
```
